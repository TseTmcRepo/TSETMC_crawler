# Fetch All Namaads Info

The resource page has 5 diffrent data structure seperated by `@`. The third part has namaads info seperated by `;`.

```python
response = requests.get('http://www.tsetmc.com/tsev2/data/MarketWatchInit.aspx')
content = response.content.decode('utf-8')
a = content.split('@')[2].split(';')
namaad_data = a[200].split(',')
namaad_data[0], namaad_data[2]
```

# Get Histry Details

```python
response = requests.get('http://cdn.tsetmc.com/Loader.aspx?ParTree=15131P&i=30852391633490755&d=20191001')

javascript_var_regex='var (\w+)=((\'|\[)[^;]+(\'|\]));'
content = response.content.decode('utf-8')
js_var_dict = {}
parsed_js_vars = findall(javascript_var_regex, content)
for m in parsed_js_vars:
    var_content = m[1].replace('\'','\"')
    js_var_dict[m[0]]=json.loads(var_content)
pprint(js_var_dict.keys())
```
